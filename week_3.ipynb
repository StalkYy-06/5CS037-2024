### What percent of the total number of passengers on the Titanic belonged to each age group

```python
# Creating a new column "age_group" that categorizes passengers into "child" (age < 18), "adult" (age 18{64), and "senior" (age 65 and above).

#setting the default value to "N/A"
df['age_group'] = 'N/A'

#changing the value to child if the age is below 18
df.loc[df['Age'] < 18, 'age_group'] = 'child'

# changing the value to adult if the age is below 65 and above 18
df.loc[(df['Age']>18) & (df['Age']<65),'age_group']='adult'

#changing the value to senior if the age is above 65
df.loc[df['Age'] >= 65, 'age_group'] = 'senior'

df.head()
```

```python
# calculating the total number of passengers on titanic

total_passengers = len(df)

total_passengers
```

```python
# counting the number of passengers in each age group

age_group_counts = df['age_group'].value_counts()

age_group_counts
```

```python
# Divide the count of each age group by the total number of passengers to get the proportion of passengers in each age group.

age_group_proportions = age_group_counts / total_passengers

age_group_proportions
```

```python
# converting the proportions to percentage

age_group_percentages = age_group_proportions * 100

# displaying the proportions as a percentage
age_group_percentages
```
